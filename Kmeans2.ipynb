```python
#step 1: import our dependencies
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

img = mpimg.imread('girl3.jpg') #now the image is just a tensorflow with a lot of 3 dimension vectors
plt.imshow(img) #to show it we have to ust plt with function imshow()
imgplot = plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
#convert the image into a matrix with each row consist 1 pixel with 3 colors
X = img.reshape((img.shape[0]*img.shape[1],img.shape[2]))
X
```

```python
for K in [2,5,10,15,20]:
    kmeans = KMeans(n_clusters = K).fit(X) #computing the k mean clustering
    label = kmeans.predict(X) # Predict the closest cluster each sample in X belongs to 
    
    img4 = np.zeros_like(X) #filling the new img4 with all zeros 
    #Actually we are creating a new frame that is empty is img4
    #replace each pixel by its center
    for k in range(K):
        img4[label == k] = kmeans.cluster_centers_[k]
    #Reshape and display the image
    img5 = img4.reshape((img.shape[0],img.shape[1],img.shape[2]))
    plt.imshow(img5, interpolation = 'nearest')
    plt.axis('off') #remove the axis
    plt.show()
```
